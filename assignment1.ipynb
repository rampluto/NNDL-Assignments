Q.1(a)Write python code for plotting the following activation functions for the input x in the range of -20
to 20.
(a) Sigmoid (b) TanH (c) Softsign (d) ReLU (e) Softmax

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-20, 20)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-20, 20)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :TanH')
plt.show()
```

```python
def sofsign(x):
    return x/(1+np.abs(x))
```

```python
x = np.linspace(-20, 20)
plt.plot(x, sofsign(x))
plt.axis('tight')
plt.title('Activation Function :Softsign')
plt.show()
```

```python
def relu(x):
    x1 =list()
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
```

```python
x = np.linspace(-20, 20)
plt.plot(x, relu(x))
plt.axis('tight')
plt.title('Activation Function :Relu')
plt.show()
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)
```

```python
x = np.linspace(-20, 20)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.show()
```

(b)Write python code for the derivative of the following activation functions and their plots for the input x
in the range of -20 to 20.
(a) Sigmoid (b) TanH (c) Softsign (d) ReLU

```python
def sigmoid_derivative(x):
    s = 1/(1+np.exp(-x))
    ds = s*(1-s)
    return ds
```

```python
x = np.linspace(-20, 20)
plt.plot(x, sigmoid_derivative(x))
plt.axis('tight')
plt.title('Derivative of Activation Function :Sigmoid')
plt.show()
```

```python
def tanh_derivative(x):
    a = np.tanh(x)
    da = 1-(a*a)
    return da
```

```python
x = np.linspace(-20, 20)
plt.plot(x, tanh_derivative(x))
plt.axis('tight')
plt.title('Derivative of Activation Function :TanH')
plt.show()
```

```python
def softsign_derivative(x):
    t = (1+np.abs(x))
    t = t**2
    return 1/t
```

```python
x = np.linspace(-20, 20)
plt.plot(x, softsign_derivative(x))
plt.axis('tight')
plt.title('Derivative of Activation Function :Softsign')
plt.show()
```

```python
def relu_derivative(x):
    ds = list()
    for i in x:
        if(i>0):
            ds.append(1)
        else:
            ds.append(0)
    return ds
```

```python
x = np.linspace(-20, 20)
plt.plot(x, relu_derivative(x))
plt.axis('tight')
plt.title('Derivative of Activation Function :Relu')
plt.show()
```

Q.3. Write python code to compute the following Regression loss value for the given true output and
predicted output by network:
y_true = [11, 20,19,17,10,24,23] ## Target or actual Value
y_pred =[12,18,19.5,18,9,23,24] ## Predicted value by ANN model
(a) MSE (b) MAE (c) MBE (d) Huber Loss (e) Epsilon Hinge Loss (d) Square Epsilon Hinge Loss
(e)verify each loss value by analytical method

```python
def mse(y_true, y_pred):
    return np.square(np.subtract(y_true,y_pred)).mean()
```

```python
y_true = [11,20,19,17,10,24,23]
y_pred = [12,18,19.5,18,9,23,24]
```

```python
print("MSE is",mse(y_true,y_pred))
```

```python
def mae(y_true,y_pred):
    return np.abs(np.subtract(y_true,y_pred)).mean()
```

```python
print("MAE is",mae(y_true,y_pred))
```

```python
def mbe(y_true,y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    y_true = y_true.reshape(len(y_true),1)
    y_pred = y_pred.reshape(len(y_pred),1)
    diff = y_true-y_pred
    mbe = diff.mean()
    return mbe
```

```python
print("MBE is",mbe(y_true,y_pred))
```

```python
import tensorflow as tf
def huber_loss(y_true,y_pred):
    hub_loss = tf.keras.losses.huber(y_true,y_pred).numpy()
    return hub_loss
```

```python
print("Huberloss is",huber_loss(y_true,y_pred))
```

```python
def epsilon_hinge_loss(y_true,y_pred):
    hinge_loss = tf.keras.losses.hinge(y_true,y_pred).numpy()
    return hinge_loss
```

```python
print("Epsilon Hinge loss is",epsilon_hinge_loss(y_true,y_pred))
```

```python
def squared_hinge_loss(y_true,y_pred):
    sqhinge_loss = tf.keras.losses.SquaredHinge()
    return sqhinge_loss(y_true,y_pred).numpy()
```

```python
print("squared Hinge loss is",squared_hinge_loss(y_true,y_pred))
```

Q4.Write python code to compute the values of following Binary classification functions for the given
true output and predicted output by network:
y_pred = [0.99, 0.11, 0.11, 0.99, 0.11, 0.11, 0.99, 0.99, 0.99, 0.11, 0.99, 0.99, 0.11, 0.99, 0.99]-Predicted value by ANN model
y_true = [1, 1, 0, 1, 0, 0, 1, 0, 0, 0, 1, 1, 0, 0, 1] ## Target or actual Value
(a) Binary cross entropy loss (b) Jaccard Loss (c) Dice loss (d) verify each loss value by analytical
method

```python
def bce_loss(y_true,y_pred):
    bce = tf.keras.losses.BinaryCrossentropy(from_logits=True)
    return bce(y_true, y_pred).numpy()
```

```python
y_pred = [0.99, 0.11, 0.11, 0.99, 0.11, 0.11, 0.99, 0.99, 0.99, 0.11, 0.99, 0.99, 0.11, 0.99, 0.99]
y_true = [1, 1, 0, 1, 0, 0, 1, 0, 0, 0, 1, 1, 0, 0, 1]
print("Binary Cross entropy loss is",bce_loss(y_true,y_pred))
```

```python
def jaccard_loss(y_true,y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    t1 = tf.reduce_sum(y_true * y_pred, axis=-1)
    t2 = tf.reduce_sum(y_true + y_pred, axis=-1)
    jac = t1 / (t2-t1)
    return tf.reduce_mean(jac).numpy()
```

```python
print("Jaccard loss is",jaccard_loss(y_true,y_pred))
```

```python
def dice_loss(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    numerator = 2 * tf.reduce_sum(y_true * y_pred, axis=-1)
    denominator = tf.reduce_sum(y_true + y_pred, axis=-1)
    return (1 - (numerator + 1) / (denominator + 1)).numpy()
```

```python
print("Dice loss is",dice_loss(y_true,y_pred))
```

Q.6.The neural network shown in Fig.1 has the following hyper parameters and input: Choose random
weights of the neuron and bias=0, learning rate =0.01 and inputs to the
neuron and target values are as follows.
Write a python code for calculating the output of neural network using
Gradient Descent algorithm

```python
#data 
x1 = [4,2,1,3,1,6]
x2 = [1,8,0,2,4,7]
y = [2,-14,1,-1,-7,-8]
```

```python
#building model y=w1x1+w2x2+b
w1 = 0
w2 = 0
b  = 0
epochs = 10
lr = 0.01
batches = len(x1)
for i in range(epochs):
    for j in range(batches):
        y_pred = w1*x1[j]+w2*x2[j]+b
        loss = (y_pred-y[j])**2
        dl = 2*(y_pred-y[j])
        db = 1
        dw1 = x1[j]
        dw2 = x2[j]
        b = b-lr*dl*db
        w1 = w1-lr*(dl*dw1)
        w2 = w2-lr*(dl*dw2)
    print("Epoch :",i+1,"/",epochs,"loss: ",loss)
print(w1,w2,b)
```

```python
#predicted value of Y finally
y_pred=[]
for i in range(len(x1)):
    yp = w1*x1[i]+w2*x2[i]+b
    y_pred.append(yp)
print(y_pred)
```

```python
sums = 0
for i in range(6):
    sums = sums+((y[i]-y_pred[i])**2)
    
print("mse is",sums/float(len(x1)))
```
